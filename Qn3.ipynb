```python
#Read any data frame and do the following plots?

#1-distplot
#2-Scatter
#3-violin
#4-Histogram
#5-Box and whisker plot
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/content/Global new2.csv")
sns.distplot(df)
```

```python
#scatter plot
plt.grid(True,color='k',linestyle=':')
plt.scatter(df.Month,df.Profit,color='g',s=100)
```

```python
#violin plot
sns.catplot(x='Month',y='Profit',data=df,kind='violin')
```

```python
#histogram
plt.grid(True,color='k',linestyle=':')
plt.hist(df)
```

```python
#Box and whisker plot
plt.grid(True,color='k',linestyle=':')
plt.boxplot(df)
```
